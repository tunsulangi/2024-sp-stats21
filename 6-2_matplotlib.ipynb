## Creating a simple plot.

First, we create arrays of the values to plot.

We create an array of 500 values from 0 to pi.

```python
x = np.linspace(0, np.pi * 2, 500)
```

We then calculate the y values we wish to plot. In this case, we'll keep it simple and calculate sin of x (in radians).

```python
y = np.sin(x)
```

# Line Plot

To create a plot, you can call `plt.plot(x, y)`. To have it appear, you call `plt.show()`. `plt.show()` is much like calling `print()`. In Jupyter, if you do not call `plt.show()`, the plot will still often appear, but it is generally considered **good practice to call `plt.show()`**.

Prior to calling `plt.show()`, you can call optionally other functions which will modify the plot, such as adding a title or changing the axis limits.

```python
plt.plot(x, y)
plt.title("Sin of x") # optional
plt.xlim(0,2 * np.pi) # optional
plt.show() 
```

The default behavior of `plt.plot()` is to plot the points (about 500, actually) and connect them with lines. 

```python
x = np.arange(0, 5, 1)
print(x)
y = np.random.randint(0, 10, 5)
print(y)
plt.plot(x, y)
plt.show()
```

Since the values generated for both the x and y are random, this plot will look different than in-class example.

# Scatterplot

If you do not want the points to be connected with a line, you can ask for a scatterplot.

```python
plt.scatter(x,y)
plt.show()
```

# Multiple functions or sequences on the same graph

Repeated calls to `plt.plot()` will add lines to the same plot.

```python
x = np.linspace(0, np.pi * 2, 500)
y = np.sin(x)
z = np.tan(x)
w = np.cos(x)
```

```python
plt.plot(x, y, 'blue')
plt.plot(x, z, 'red')
plt.show() # the scale of the two functions make this graph undesirable
```

`sin(x)` gets flattened out because tangent goes to infinities. The scale doesn't actually go to infinities.

```python
plt.plot(x, y, 'blue')
plt.plot(x, w, 'orange')
plt.show()
```

Note the buffer between the x-axis and max of the graphs.

# changing labels and axis limits

```python
plt.plot(x, y)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

Y-axis goes from -1 to 1, so the scale for that differs than x-axis.

```python
plt.plot(x, y)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(0, 2 * np.pi)
plt.axis('equal')  # aspect ratio # investigate later
plt.show()
```

With `plt.axis('equal')`, if there was a square or circle graphed, they would look like they should, instead of squished or squashed.

```python
plt.scatter(x, y, s = 300, marker = 'd', color = 'lavender')
plt.scatter(x, w, s = 200, marker = 'H', color = 'darkslategrey')
plt.xlim([0, .2])
plt.show()
```

# Plot options: figure size, axis, title

Note that `plt.figure(figsize = (6, 4))` specifies the size **before** creating the actual plots.

```python
np.random.seed(1)
x = np.random.random(100)
y = np.random.random(100)
plt.figure(figsize = (6,4))  # define the properties of the figure first
plt.scatter(x,y)             # then add content
plt.axis('equal')            # alter properties
plt.title('2-dimensional random uniform values')
plt.show()
```

# Subplots example

```python
x = np.linspace(0, np.pi * 2, 500)
y = np.sin(x)
z = np.tan(x)
w = np.cos(x)
plt.subplot(2,2,1)  # number of rows, number of columns, which plot you want to draw
plt.plot(x, y)
plt.title("sin(x)")

plt.subplot(2,2,2) # subplot 2 will be upper right
plt.plot(x, w)
plt.title("cos(x)")

plt.subplot(2,2,3) # subplot 3 will be lower left
plt.plot(x, z)
plt.title("tan(x)")


plt.subplot(2,2,4)
plt.plot(x,y, 'red')
plt.plot(x,w, 'blue')
plt.title("sin and cos")

plt.show()
```

The different code blocks above designate the different plots on the 2x2 area.

```python
plt.figure(figsize = (10,5))
# More Complicated Grids
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
plt.subplot(grid[0, 0]) # gridspec uses 0 based indexing
plt.scatter(x[0:10], y[0:10])
plt.subplot(grid[0, 1:3])  # top row, columns 1:3
plt.hist(x)
plt.subplot(grid[1, :2])
plt.hist(y)
plt.subplot(grid[1, 2])
plt.plot(x[0:4], y[0:4], c = 'r')
plt.show()
```

Note the `'stepfilled'` argument below and the difference of that and the above plot.

```python
# histtype = stepfilled
plt.hist(data, bins=25, alpha=1,
         histtype = 'stepfilled', color='steelblue', edgecolor='black')
plt.show()
```

hist documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

It's good to just set `kwargs` once below so if you want to change, say the bins, it would automatically change it for each plot that uses it.

```python
x1 = np.random.normal(0, 0.8, 10000)
x2 = np.random.normal(-2, 1, 10000)
x3 = np.random.normal(3, 2, 10000)

# set up a dictionary with arguments
kwargs = dict(histtype='stepfilled', alpha=0.6, density=True, bins=20, edgecolor = 'black')

# use the same arguments for all of the histograms without the need to copy and paste
plt.hist(x1, **kwargs, color = 'navy') # scattering **kwargs, like copy pasting the dictionary
plt.hist(x2, **kwargs, color = "purple")
plt.hist(x3, **kwargs)
plt.show()
```

```python
# 2d histograms
mean = [0, 0]
cov = [[1, 1], 
       [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
plt.show()
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

## two dimensional functions

```python
x = np.array([0,1,2,3])
y = np.array([-1,0,1, 2])
print(x)
print(y)
X, Y = np.meshgrid(x,y) # sees x has length 3, y has length 4, so the result is a 4 x 3 grid
```

Broadcasting above, which can be seen when printing out X and Y below.

```python
print(X)
print(Y)
```

```python
def f(x, y):
    return x * y
X, Y = np.meshgrid(x,y)
Z = f(X, Y)
print(Z)
```

The product of X and Y, element-wise. E.g 0 * -1 = 0

```python
x = np.linspace(-3, 3, 101)
y = np.linspace(-3, 3, 101)
X, Y = np.meshgrid(x, y)

def g (x, y): 
    return np.exp(-(x**2)) * np.exp(-(y**2)) # product of 2 normal distributions

Z = g(X, Y)
```

```python
Z.shape
```

```python
plt.contour(X, Y, Z)
plt.axis('equal')
plt.show()
```

Contour plot from top-down

```python
plt.imshow(Z)
plt.show()
```

### Titanic

```python
# Titanic Dataset
url = 'https://assets.datacamp.com/production/course_1607/datasets/titanic_sub.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python
titanic.Survived.value_counts().plot(kind = "bar")
```

```python
titanic.Survived.value_counts()
```

```python
titanic.dtypes
```

Survived is just 0 or 1 for each passenger.

```python
titanic.Survived = titanic.Survived.astype("category")
```

```python
titanic.dtypes
```

Now it's categorical data. As such it can be renamed.

```python
titanic.Survived = titanic.Survived.cat.rename_categories({0: "died", 1:"survived"})
```

The 0's are now replaced with died, and 1's with survived.

```python
pd.crosstab(titanic.Sex, titanic.Survived)
```

```python
pd.crosstab(titanic.Sex, titanic.Survived).plot(kind = "bar", stacked = True)
```

Great discrepancy. Women got put on the lifeboats vs the men.

```python
titanic["is_child"] = titanic.Age < 8
```

```python
pd.crosstab(titanic.is_child, titanic.Survived)
```

Proportions can be got (below)

```python
pd.crosstab(titanic.is_child, titanic.Survived, normalize = 0)
```

`normalize = 1` -> normalize across the columns instead of rows

```python
pd.crosstab(titanic.is_child, titanic.Survived, normalize = 0).plot(kind = "bar", stacked = True)
```

## Groupby

```python
titanic_summary = titanic.groupby(["Sex", "is_child", "Survived"], observed = False)["PassengerId"].count()
```

```python
titanic_summary
```

female adult died and survived counts, then female child died and survived counts, etc.

```python
titanic_summary.unstack()
```

```python
titanic_summary.unstack().plot(kind = "bar", stacked = True)
```

No normalize function directly for this case.

```python
titanic_summary.unstack().sum(axis = 1)
```

```python
row_totals = titanic_summary.unstack().sum(axis = 1)
```

```python
titanic_summary.unstack().div(row_totals, axis = 0)
```

Now we have the proportions (above)

```python
titanic_summary.unstack().div(row_totals, axis = 0).plot(kind = "bar", stacked = True)
```

83% adult males died

#### Create a data frame consisting of random die rolls

Column 'one' shows the sum after one roll. (min is 1, max is 6, dist is uniform)

Column 'two' shows the sum after two rolls. (min is 2, max is 12, dist is triangular)

```python
df = pd.DataFrame(np.random.randint(1, 7, 60000), columns = ['one'])
df['two'] = df['one'] + np.random.randint(1, 7, 60000)
ax = df.plot.hist(bins=12, alpha=0.7)
```

Column 'three' shows the sum after three rolls. (min is 3, max is 18)

And so on... as we add more dice rolls, we see the central limit theorem start to take effect.

```python
df['three'] = df['two'] + np.random.randint(1, 7, 60000)
df['four'] = df['three'] + np.random.randint(1, 7, 60000)
df['five'] = df['four'] + np.random.randint(1, 7, 60000)
```

```python
df.plot.hist(bins = 30, alpha = 0.5)
```
